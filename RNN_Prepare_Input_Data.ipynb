```python
import numpy as np
```

```python
docs = ['go nepal',
        'nepal nepal',
        'hip hip huray',
        'mountering country nepal',
        'balen balen',
        'balen ko jay',
        'prime minister of nepal is nepal']
```

## Step 1: Tokenize data:
Tokanize means break down sentences into word and convert capital sentences into small letters

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer(oov_token='nothing')  # oov_token='nothing' means incase of not getting the words from the documents while predicating the words appear 'nothing'
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences = tokenizer.texts_to_sequences(docs)
```

```python
sequences
```

```python
from tensorflow.keras.utils import pad_sequences
```

```python
sequences = pad_sequences(sequences, padding='post')
sequences
```
